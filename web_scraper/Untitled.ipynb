```python
df_products = pd.read_csv(import_path_products)
```

```python
ingredients_category = ['Fresh Fruits & Veggies','Meat, Seafood & Plant-Based', 'Cheese', 'Dairy & Eggs']
```

```python
df_filtered_products = df_products[df_products['category'].isin(ingredients_category)]
```

```python
df_recipes = pd.read_csv(import_path_recipes)
```

```python
df_recipes_filtered = df_recipes.drop(columns = ['url','image_url'])
```

```python
df_recipes_filtered.sample(10)
```

```python
import re

def normalize(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)  # remove punctuation
    text = text.replace("trader joes", "tj")
    text = text.replace("tjs", "tj")
    text = re.sub(r"\btj\b", "trader joes", text)  # unify abbreviations
    return text.strip()
```

```python
df_recipes_filtered['ingredients'].iloc[0]
```

```python
df_recipes_filtered['ingredients_norm'] = df_recipes_filtered['ingredients'].apply(
    lambda lst: [i for i in lst] if isinstance(lst, list) else []
)
df_filtered_products['product_norm'] = df_filtered_products['product_name'].apply(normalize)
```

```python
from tqdm import tqdm

product_names = df_filtered_products[['product_name', 'product_norm']].values.tolist()

matches = []
for _, recipe in tqdm(df_recipes_filtered.iterrows(), total=len(df_recipes_filtered)):
    recipe_matches = []
    for ingredient in recipe['ingredients_norm']:
        for product_name, product_norm in product_names:
            if product_norm in ingredient:
                recipe_matches.append(product_name)
    matches.append(list(set(recipe_matches)))

df_recipes_filtered['matched_products'] = matches

```
